# Removal of Redundant Columns

The following columns are redundant for predicting the output class:
- **obj_ID**: This is a unique identifier for each object in the image catalog, but it does not provide any information about the object’s properties or characteristics that could help classify it as a star, galaxy or quasar.
- **run_ID**: This is a number that identifies the specific scan of the sky, but it does not affect the object’s appearance or features that could help classify it as a star, galaxy or quasar.  
- **rereun_ID**: This is a number that specifies how the image was processed, but it does not change the object’s intrinsic properties or characteristics that could help classify it as a star, galaxy or quasar.  
- **cam_col**: This is a number that identifies the scanline within the run, but it does not influence the object’s shape or color that could help classify it as a star, galaxy or quasar.  
- **field_ID**: This is a number that identifies each field of view in the sky, but it does not determine the object’s location or distance that could help classify it as a star, galaxy or quasar.  
- **spec_obj_ID**: This is a unique identifier for each object that has optical spectroscopic data, but it does not reveal any information about the object’s spectrum or composition that could help classify it as a star, galaxy or quasar.   
- **MJD**: This is a date that indicates when a given piece of SDSS data was taken, but it does not reflect any changes in the object’s brightness or motion that could help classify it as a star, galaxy or quasar. 
- **fiber_ID**: This is a number that identifies the fiber that pointed the light at the focal plane in each observation, but it does not alter the object’s shape or size that could help classify it as a star, galaxy or quasar.  
Therefore, we would say that these 9 columns are redundant and can be dropped from the data without affecting the prediction of the output class.  
You should consult with an expert before making any decisions based on data analysis. In this case, domain expertise is necessary


```python
# Assuming that the dataframe df has already been imported from the csv file

# Creating a list of the redundant columns to be checked

redundant_columns = ["obj_ID", "run_ID", "rerun_ID", "cam_col", "field_ID", "spec_obj_ID",  "MJD", "fiber_ID"]
```

```python

# Creating a new dataframe with only the redundant columns and the output class
# To analyse correlation between the redundant columns and the output class
df_redundant = df[redundant_columns + ["class"]]
# df_redundant.head()

# Now we will check the correlation between the redundant columns and the output class
df_redundant.corr()

# Use heatmap to visualise the correlation
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(df_redundant.corr(), annot=True, fmt='.0%')
plt.show()

```

```python
# Assuming that the dataframe df has already been imported from the csv file

# Dropping the redundant columns from the dataframe and assigning the result to a new dataframe df_clean
df_clean = df.drop(columns=redundant_columns)

# Printing the first 5 rows of the new dataframe to check the result
print(df_clean.head())
```

```python
# Without Removing the Redundant Columns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Splitting the data into features (X) and output class (y)
X = df.drop(columns="class")
y = df["class"]

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=35)

# Creating and fitting a logistic regression model on the training data
model = LogisticRegression(max_iter=800)
model.fit(X_train, y_train)

# Making predictions on the testing data
y_pred = model.predict(X_test)

# Calculating and printing the accuracy score of the model
accuracy = accuracy_score(y_test, y_pred)
print("The accuracy of the model is:", accuracy)
```

```python
# After Removing the Redundant Columns
x = df_clean.drop(columns="class")
y = df_clean["class"]

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=35)

# Creating and fitting a logistic regression model on the training data, with max_iter=1000
model = LogisticRegression(max_iter=800)
model.fit(x_train, y_train)

# Making predictions on the testing data
y_pred = model.predict(x_test)

# Calculating and printing the accuracy score of the model
accuracy = accuracy_score(y_test, y_pred)
print("The accuracy of the model is:", accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier

x = df_clean.drop(columns="class")
y = df_clean["class"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=35)


rfc = RandomForestClassifier(n_estimators=100, random_state=35)
rfc.fit(x_train, y_train)

# Making predictions on the testing data
y_pred = rfc.predict(x_test)

# Calculating and printing the accuracy score of the model
accuracy = accuracy_score(y_test, y_pred)
print("The accuracy of the model is:", accuracy)
```
